### Task 1: Trends in Compensation Over Age Groups


##### 1. Line Chart of Median `ConvertedCompYearly` by Age Group


- Track how the median yearly compensation (ConvertedCompYearly) changes across different age groups.

- Use a line chart to visualize these trends.


```python
# Ensure the data is clean, especially the 'ConvertedCompYearly' column
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

# Remove rows with NaN values in 'ConvertedCompYearly' and 'Age'
df_cleaned = df.dropna(subset=['ConvertedCompYearly', 'Age'])

# Calculate the median compensation for each age
median_comp_by_age = df_cleaned.groupby('Age')['ConvertedCompYearly'].median()

# Sort by age for a meaningful line chart
median_comp_by_age = median_comp_by_age.sort_index()

# Plot the line chart
plt.figure(figsize=(8, 6))
plt.plot(median_comp_by_age.index, median_comp_by_age.values)
plt.title('Median Yearly Compensation by Age')
plt.xlabel('Age')
plt.ylabel('Median ConvertedCompYearly')
plt.xticks(rotation=90)
plt.grid(True)
plt.show()
```

##### 2. Line Chart of Median `ConvertedCompYearly` for Ages 25 to 45


For a closer look, plot a line chart focusing on the median compensation for respondents between ages 25 and 45.


```python
age_groups = ['25-34 years old', '35-44 years old']
df_filtered = df[df['Age'].isin(age_groups)]

# Calculate the median compensation for each age group
median_compensation = df_filtered.groupby('Age')['ConvertedCompYearly'].median().reset_index()

# Plot the line chart
plt.figure(figsize=(8, 6))
plt.plot(median_compensation['Age'], median_compensation['ConvertedCompYearly'], marker='o', linestyle='-')
plt.title('Median Compensation for Developers Aged 25-45')
plt.xlabel('Age Group')
plt.ylabel('Median Compensation (Yearly)')
plt.grid(True)
plt.show()
```

### Task 2: Trends in Job Satisfaction by Experience Level



##### 1. Line Chart of Job Satisfaction (`JobSatPoints_6`) by Experience Level



- Use a column that approximates experience level to analyze how job satisfaction changes with experience.

- If needed, substitute an available experience-related column for `Experience`.


```python
df['JobSatPoints_6'] = pd.to_numeric(df['JobSatPoints_6'], errors='coerce')

# Convert 'YearsCode' to numeric, handling potential string values like 'More than 50 years' and 'Less than 1 year'
def clean_years_code(yc):
    if yc == 'More than 50 years':
        return 51  # Approximate with a value slightly above 50
    elif yc == 'Less than 1 year':
        return 0.5 # Approximate with a value between 0 and 1
    try:
        return int(yc)
    except:
        return None # Handle other non-numeric cases

df['YearsCode_numeric'] = df['YearsCode'].apply(clean_years_code)

# Remove rows with NaN values in 'JobSatPoints_6' and 'YearsCode_numeric'
df_cleaned_exp = df.dropna(subset=['JobSatPoints_6', 'YearsCode_numeric'])

# Calculate the median job satisfaction for each year of coding experience
median_jobsat_by_experience = df_cleaned_exp.groupby('YearsCode_numeric')['JobSatPoints_6'].median()

# Sort by experience level for a meaningful line chart
median_jobsat_by_experience = median_jobsat_by_experience.sort_index()

# Plot the line chart
plt.figure(figsize=(8, 6))
plt.plot(median_jobsat_by_experience.index, median_jobsat_by_experience.values)
plt.title('Median Job Satisfaction by Years of Coding Experience')
plt.xlabel('Years of Coding Experience (Approximation of Experience Level)')
plt.ylabel('Median Job Satisfaction (JobSatPoints_6)')
plt.grid(True)
plt.show()
```

### Task 3: Trends in Job Satisfaction and Compensation by Experience


##### 1.Line Chart of Median ConvertedCompYearly Over Experience Level

- This line chart will track how median compensation (`ConvertedCompYearly`) changes with increasing experience.

- Use a column such as `WorkExp` or another relevant experience-related column.


```python
# Convert 'ConvertedCompYearly' to numeric, handling potential errors
df['ConvertedCompYearly'] = pd.to_numeric(df['ConvertedCompYearly'], errors='coerce')

# Use 'WorkExp' column for experience level. Convert to numeric, handling potential errors.
df['WorkExp_numeric'] = pd.to_numeric(df['WorkExp'], errors='coerce')

# Remove rows with NaN values in 'ConvertedCompYearly' and 'WorkExp_numeric'
df_cleaned_comp_exp = df.dropna(subset=['ConvertedCompYearly', 'WorkExp_numeric'])

# Calculate the median compensation for each work experience level
median_comp_by_workexp = df_cleaned_comp_exp.groupby('WorkExp_numeric')['ConvertedCompYearly'].median()

# Sort by experience level for a meaningful line chart
median_comp_by_workexp = median_comp_by_workexp.sort_index()

# Plot the line chart
plt.figure(figsize=(8, 6))
plt.plot(median_comp_by_workexp.index, median_comp_by_workexp.values)
plt.title('Median ConvertedCompYearly Over Work Experience Level')
plt.xlabel('Work Experience Level (Years)')
plt.ylabel('Median ConvertedCompYearly')
plt.grid(True)
plt.show()
```

##### 2.Line Chart of Job Satisfaction (`JobSatPoints_6`) Across Experience Levels

- Create a line chart to explore trends in job satisfaction (`JobSatPoints_6`) based on experience level.

- This chart will provide insight into how satisfaction correlates with experience over time


```python
# Calculate the median job satisfaction for each work experience level
median_jobsat_by_workexp = df_cleaned_comp_exp.groupby('WorkExp_numeric')['JobSatPoints_6'].median()

# Sort by experience level for a meaningful line chart
median_jobsat_by_workexp = median_jobsat_by_workexp.sort_index()

# Plot the line chart
plt.figure(figsize=(8, 6))
plt.plot(median_jobsat_by_workexp.index, median_jobsat_by_workexp.values)
plt.title('Median Job Satisfaction Over Work Experience Level')
plt.xlabel('Work Experience Level (Years)')
plt.ylabel('Median Job Satisfaction (JobSatPoints_6)')
plt.grid(True)
plt.show()
```
